# Load libraries and video file

```python
import cv2 # utilize cv functions
import numpy as np # matrix manipulations
import time # delay video 

# Capture input video
cap = cv2.VideoCapture('traffic.mp4')
```

# Threshold =10
# Create Back subtractorMoG2 model

```python
# Background subtractor model created 
fgbg = cv2.createBackgroundSubtractorMOG2(history=100, varThreshold=10, detectShadows=True)

#Paramenters
length_min = 80 # Minimum length of the retangle
height_min = 80 # Minimum height of the rectangle

offset = 6 # Error allowed between pixel

line_pos = 200 # Count line position

delay = 60 # FPS of video

detect = []
cars = 0
while(1):
    #read video frame
    ret, frame = cap.read()
    #look for changes
    fgmask = fgbg.apply(frame)
    # Add delay
    time.sleep(.02)
    
    # computes the bounding box for the contour, and draws it on the frame
    contour, h = cv2.findContours(fgmask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    font = cv2.FONT_HERSHEY_SIMPLEX
    #cv2.line(frame1, (80, line_pos), (705, line_pos), (0,255,0), 5)
    
    for (i,c) in enumerate (contour):
        (x, y, w, h) = cv2.boundingRect(c)
        validate_contour = (w >= length_min) and (h >= height_min)
        if not validate_contour:
            continue
        for (x,y) in detect:
            if y<(lin_pos+offset) and y>(lin_pos-offset):
                cars+=1
                cv2.line(frame, (80, lin_pos), (720, lin_pos), (0,0,255), 10)
                cv2.line(frame, (20, lin_pos), (605, lin_pos), (0,0,255), 10)
                detect.remove((x,y))

        # draw the contour and center of the shape on the image
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0,255,0), 2)
        print("Vehicles detected: "+str(cars))        

    #cv2.putText(frame, "Vehicles Detected: "+str(cars), (5, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255),3)
    #cv2.putText(frame, "ROI "+str(cars), (600, 190), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255),2)
    #cv2.putText(frame, str(cap.get(cv2.CAP_PROP_POS_FRAMES)), (15, 15),
               #cv2.FONT_HERSHEY_SIMPLEX, 0.5 , (0,0,0))
    # Show window and all frame 
    cv2.imshow('fgmask  threshold=10',fgmask)
    cv2.imshow('frame',frame)
    
    # Press Esc key to stop
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break

# Clean up everything when finished 
cap.release()
cv2.destroyAllWindows()
```
